```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import random

# Path to the directory containing subdirectories with images
main_dir = '/kaggle/input/new-bangladeshi-crop-disease/BangladeshiCrops/BangladeshiCrops/Crop___Disease/Corn'

# Get the list of subdirectories (folders) in the main directory
sub_dirs = os.listdir(main_dir)

# Initialize a counter for plotted images
plot_count = 0

# Initialize a figure to plot the images
plt.figure(figsize=(12, 12))

# Loop through each subdirectory
for sub_dir in sub_dirs:
    # Get the path to the current subdirectory
    sub_dir_path = os.path.join(main_dir, sub_dir)
    
    # Get the list of files (images) in the current subdirectory
    files = os.listdir(sub_dir_path)
    
    # Choose a random image from the files list
    random_image = random.choice(files)
    
    # Get the path to the random image
    image_path = os.path.join(sub_dir_path, random_image)
    
    # Open the image using PIL
    img = Image.open(image_path)
    
    # Plot the image
    plot_count += 1
    plt.subplot(5, 2, plot_count)  # Adjust the subplot layout as needed
    plt.imshow(img)
    plt.title(sub_dir)
    plt.axis('off')
    
    # Stop plotting after 10 images
    if plot_count == 4:
        break

# Adjust the layout and show the plot
plt.tight_layout()
plt.show()
```
